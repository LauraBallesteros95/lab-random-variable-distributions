# Random variable distributions

```python
#Step 1. Import the libraries:
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import warnings
import matplotlib.pyplot as plt 
pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv(r'C:\Users\laura\Desktop\ironhack\MySQL\lab-random-variable-distributions\files_for_lab\we_fn_use_c_marketing_customer_value_analysis.csv')
```

- Step 1. Get the numerical variables from our dataset.

```python
numerical= df.select_dtypes(include= np.number)
```

- Step 2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

```python
#Use a distribution plot to see if the variables are normally distributed
for col in numerical.columns:
    sns.distplot(numerical[col], color='purple')
    plt.show()
```

### Normally Distributions:
- We only have two columns with normaly distributed: 
    - Customer lifetime value
    - Months since policy inception

```python
#Apply the logaritm to the variables to see if they're normally distributed
for col in numerical.columns:
    numerical[col]=np.log(numerical[col]+1)
    sns.distplot(numerical[col], color='pink')
    plt.show()
```

By surprise, having done the logistic transformation, the distributions have been disordered and it is no longer the original normal distribution, therefore

- Step 3Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
#Apply the logaritm to the variables to see if they're normally distributed
for col in numerical.columns:
    numerical[col]=np.sqrt(numerical[col])
    sns.distplot(numerical[col], color='red')
    plt.show()
```

With the square root transformation the variables are more uniform, but finer

Finally, the client's lifetime is also no longer normally distributed.
